```python
import pandas as pd

url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
datos = pd.read_json(url)

datos.head()
```

#ðŸ”§ TransformaciÃ³n

```python
customer_datos = pd.json_normalize(datos['customer'])
customer_datos
```

```python
phone_datos = pd.json_normalize(datos['phone'])
phone_datos
```

```python
internet_datos = pd.json_normalize(datos['internet'])
internet_datos
```

```python
account_datos = pd.json_normalize(datos['account'])
account_datos
```

```python
df_datos = pd.concat([
    datos[['customerID', 'Churn']],
    customer_datos,
    phone_datos,
    internet_datos,
    account_datos
], axis=1)
```

```python
df_datos
```

```python
df_datos.info()
```

```python
df_datos.dtypes
```

```python
df_datos['Contract']  = df_datos['Contract'].str.replace('-', ' ', regex=False)
```

```python
df_datos
```

```python
df_datos.columns = df_datos.columns.str.replace('.','_', regex=False).str.lower()
df_datos
```

```python
import numpy as np
df_datos['charges_total'] = df_datos['charges_total'].replace(' ', np.nan)
df_datos['charges_total'] = df_datos['charges_total'].astype(np.float64)
df_datos
```

```python
df_datos.info()
```

```python
columnas_churn =[
    'customerid',
    'churn',
    'gender',
    'phoneservice',
    'multiplelines',
    'internetservice',
    'contract',
    'paymentmethod',
    'charges_monthly',
    'charges_total'
]

datos_df = df_datos[columnas_churn].copy()
```

```python
datos_df
```

```python
datos_df['cuentas_diarias'] = np.divide(datos_df['charges_monthly'], 30)
datos_df[['charges_monthly', 'cuentas_diarias']].head()
```

```python
datos_df.head()
```

```python
datos_df.groupby('churn')['cuentas_diarias'].mean()
```

```python
datos_df['churn_binario'] = datos_df['churn'].map({'Yes':1,'No':0})
datos_df.head()
```
